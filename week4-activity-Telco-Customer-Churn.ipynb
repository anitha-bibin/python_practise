```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
telco_data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
df=pd.DataFrame(telco_data)
df.head()
```

# 1) Compare churn count with respect to gender. 

```python
plt.figure(figsize=(10.50, 7.50))
ax=sns.countplot(telco_data['gender'],hue=telco_data['Churn'])

for container in ax.containers:
    ax.bar_label(container)
    
plt.xlabel("Gender")
plt.ylabel("Churn count")
```

# 2) Find out how many female senior citizens there in the dataset 

```python
femaleSeniorCitz=telco_data[['gender','customerID','SeniorCitizen']] [(telco_data['gender'] == 'Female') & (telco_data['SeniorCitizen'] == 1)]
print("Female Senior Citizens in the dataset is",len(femaleSeniorCitz))
```

# 3) Compare 'tenure' with 'Total Charges' 

```python
fig, ax = plt.subplots(figsize=(18,10))   
sns.stripplot(df.tenure, df.TotalCharges, jitter=0.25, size=5, ax=ax, linewidth=.5)
plt.title("Compare 'tenure' with 'Total Charges'", fontsize=20)
plt.xlabel('Tenure')
plt.ylabel("Total Charges")
plt.xticks(rotation=90)
plt.show()
```

# 4) Find out which contract preffered by the senior citizen. 

```python
#prefer=df['Contract'].value_counts(ascending=True)
#max_index = prefer.idxmax() to find the index with maximum value
#print(prefer.idxmax(), "contract preffered by the",max(prefer) ," senior citizen.")

plt.figure(figsize=(7.50, 5.50))
plt.title("contract preffered by the senior citizen")
SeniorCitzPay=telco_data[['Contract','SeniorCitizen']] [(telco_data['SeniorCitizen'] == 1)]
ax=sns.countplot(SeniorCitzPay['Contract'])
for container in ax.containers:
    ax.bar_label(container)
```
