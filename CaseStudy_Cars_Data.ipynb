# Read the dataset to python environment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cars_data=pd.read_csv('cars_data.csv')
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
cars_data.head()
```

# Check for the null values present in the dataset.

```python
#cars_data.isna()
#cars_data.info()
cars_data.isnull().sum()
```

# Plot a bar graph of male vs female buyers participated in the sales.

```python
cars_data.fillna('')
ax = cars_data['Buyer Gender'].value_counts().plot(kind='bar',figsize=(10,8),title="Male vs female buyers")
ax.set_xlabel("Count of the buyers")
ax.set_ylabel("Buyer gender")
ax.set_ylim(bottom=4000,top=5200)
ax.bar_label(ax.containers[0]) #label in bar chart in pandas
plt.show()
```

# 4. Find the top 5 cars based on their sales price.


```python
df=pd.DataFrame(cars_data)
new_data=df.nlargest(5,columns=['Sale Price'])
new_data
```

```python
#TO SELECT NECESSARY COLUMN DATA FROM DATAFRAME

df=pd.DataFrame(cars_data)
new_data=df.nlargest(5,columns=['Sale Price'])
top_list=[]
j=0
indexes=new_data.index
for i in indexes:
    top_list.insert(j,df.loc[i,["Make","Model","Sale Price"]])
    j+=1

print("Top 5 cars based on their Resell price")

for i in top_list:
    print(i["Make"]+" "+i["Model"]+" ",i["Sale Price"],"\n")
```

# 5. Find the least 5 cars based on their Resell price.

```python
df=pd.DataFrame(cars_data)
new_data1=df.nsmallest(5,columns=['Resell Price'])
new_data1
```

```python
#TO SELECT NECESSARY COLUMN DATA FROM DATAFRAME

least_list=[]
j=0
indexes=new_data1.index
for i in indexes:
    least_list.insert(j,df.loc[i,["Make","Model","Sale Price","Resell Price"]])
    j+=1

print("least 5 cars based on their Resell price")
for i in least_list:
    print(i["Make"]+" "+i["Model"]+" ",i["Sale Price"]," ",i["Resell Price"],"\n")
```
