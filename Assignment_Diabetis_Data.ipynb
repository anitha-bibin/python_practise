# Read the dataset to python environment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
diabetis_data=pd.read_excel('diabetes_model.xlsx')
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

# Display the columns in the dataset.

```python
diabetis_data.head()
```

# 1. Create a Count plot of Outcome 

```python
sns.countplot(diabetis_data['Outcome'])
```

# 2. Analyse the effect of age on Outcome using Boxplot

```python
plt.figure(figsize=(10.50, 7.50))
sns.set_style("whitegrid")
plt.ylim(10, 80)
plt.ylabel("Age")
plt.xlabel("Outcome")
sns.boxplot(x="Outcome",y="Age",data=diabetis_data,showfliers = False)
plt.title("Effect of age on outcome")
plt.show()
plt.tight_layout()
```

# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function 

```python
plt.figure(figsize=(8.5,8.5))
#sns.heatmap(diabetis_data.corr()[['DiabetesPedigreeFunction','Outcome']],annot=True)
#sns.relplot(y="DiabetesPedigreeFunction", x="Outcome", data=diabetis_data ,style="Outcome",hue="DiabetesPedigreeFunction")
sns.swarmplot(diabetis_data["Outcome"],diabetis_data["DiabetesPedigreeFunction"],s=5)
```

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome. 

```python
plt.figure()
sns.relplot(x="Glucose", y="Insulin", hue="Outcome", data=diabetis_data ,style="Outcome",markers=True,col="Outcome")
plt.xticks(rotation=75)
```

# 5. Create a heatmap on Correlation of features in the data.

```python
plt.figure(figsize=(8.5,8.5))
sns.heatmap(diabetis_data.corr(),annot=True)
plt.xticks(rotation=75)
```
